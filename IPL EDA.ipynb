From above observations ,some Data cleaning is required

1) Pune was represented by various Team Names as 'Rising Pune Supergiant', 'Pune Warriors' & 'Rising Pune Supergiants' so as a convinience we will changes these with the recent team representing Pune 'Rising Pune Supergiant' in all columns involving this name i.e 'team1','team2','winner' & 'toss_winner' columns.

2) Similarly 2nd Change is in team name of Delhi ,Earlier the team name for delhi was 'Delhi Daredevils' but later it was changed to 'Delhi Capitals' so we will replace the "delhi Daredevils' with 'Delhi Capitals'.

3) Bangalore was Renamed as Bengaluru in 2014 so we will change the Name for City Bangalore to Bengaluru to avoid Errors in Data Analysis.

```python
# To accomplish the above mentioned cleaning we will use .replace() method.
ipl_df.team1.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
ipl_df.team2.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
ipl_df.toss_winner.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
ipl_df.winner.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
ipl_df.city.replace({'Bangalore' : 'Bengaluru'},inplace=True)
```

## <font color='Orange'>Team Performance Analysis


```python
# First we'll extract our required data 
winner_df = ipl_df.groupby('winner')[['id']].count()
# sort this table in descending order
winner_df = winner_df.sort_values('id', ascending=False).reset_index()
# remove the data of 'Draw/Cancelled' bcoz it does not represent any team
winner_df=winner_df[:][winner_df['winner']!='Draw/Cancelled']
```

## <font color='DeepSkyBlue'>Player of the Match Analysis

```python
# Count the occurrences of "Player of the Match" awards
player_of_match_count = ipl_df['player_of_match'].value_counts()

# Select top N players for analysis
top_n = 10
top_players = player_of_match_count[:top_n]

# Create a bar plot to visualize the most frequently awarded players
plt.figure(figsize=(12, 8))
sns.barplot(x=top_players.index, y=top_players.values, palette="viridis")
plt.title(f'Top {top_n} Players of the Match')
plt.xlabel('Player Name')
plt.ylabel('Number of "Player of the Match" Awards')
plt.xticks(rotation=45)
plt.show()
```

## <font color = 'NavajoWhite'>Venue Impact on Match Outcome

```python
# Calculate the win percentage for each team at different venues
venue_win_percentage = ipl_df.groupby(['venue', 'winner']).size().unstack()
venue_win_percentage = (venue_win_percentage.div(venue_win_percentage.sum(axis=1), axis=0) * 100).fillna(0)

# Select a few top teams for analysis
top_teams = ['Mumbai Indians', 'Chennai Super Kings', 'Kolkata Knight Riders', 'Royal Challengers Bangalore']

# Filter data for top teams
venue_win_percentage = venue_win_percentage[top_teams]

# Create a heatmap to visualize the impact of venues on match outcomes
plt.figure(figsize=(12, 8))
sns.heatmap(data=venue_win_percentage, cmap='YlGnBu', annot=True, fmt=".1f", cbar_kws={'label': 'Win Percentage (%)'})
plt.title('Venue Impact on Match Outcomes')
plt.xlabel('Team')
plt.ylabel('Venue')
plt.xticks(rotation=45)
plt.show()
```

## <font color='	Chocolate'>Team Rivalry Analysis

```python
# Create a DataFrame to store head-to-head match counts
team_rivalries = pd.crosstab(ipl_df['team1'], ipl_df['team2'])

# Iterate through the DataFrame and calculate total head-to-head matches
for team in team_rivalries.index:
    for opponent in team_rivalries.columns:
        if team != opponent:
            total_matches = team_rivalries.at[team, opponent] + team_rivalries.at[opponent, team]
            team_rivalries.at[team, opponent] = total_matches

# Create a heatmap to visualize team rivalries
plt.figure(figsize=(12, 8))
sns.heatmap(data=team_rivalries, cmap='YlGnBu', annot=True, fmt="d", cbar_kws={'label': 'Number of Matches'})
plt.title('Team Rivalry Analysis')
plt.xlabel('Team')
plt.ylabel('Opponent')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
